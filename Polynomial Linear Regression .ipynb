```python
#Polynomial Regression 
'''
The purpose of this program is to show the advantage of Polynomial Regression over Linear Regression when the data is
non-linear. 
'''

#Importing the Libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the Dataset
loadDataset = pd.read_csv("Downloads/Position_Data.csv")
loadDataset.head()
```

```python
#Spliting the Data into X and y 
X = loadDataset.iloc[:,1:2].values
y = loadDataset.iloc[:,2].values
```

```python
#Creating Linear Regression Model 
from sklearn.linear_model import LinearRegression
linearRegressionModel = LinearRegression()
linearRegressionModel.fit(X,y)
```

```python
#Creating Polynomial Regression 
from sklearn.preprocessing import PolynomialFeatures
polynomialRegression = PolynomialFeatures(degree = 5)
polynomialRegressionModel = polynomialRegression.fit_transform(X)

polyLinearRegression = LinearRegression()
polyLinearRegression.fit(polynomialRegressionModel,y)
```

```python
#Visualizing results from Simple Linear Regression
plt.scatter(X,y,color = 'red')
plt.plot(X,linearRegressionModel.predict(X),color = 'green')
plt.title('Linear Regression')
plt.xlabel('Level of Positions')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualizing results from Polynomial Linear Regression
plt.scatter(X,y, color = 'red')
plt.plot(X,polyLinearRegression.predict(polynomialRegressionModel),color = 'green')
plt.title('Polynomial Linear Regression')
plt.xlabel('Level of Positions')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
k = np.array([9])
j = k.reshape(1,-1)
linearRegressionModel.predict(j)
```
